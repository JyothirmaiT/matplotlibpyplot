```python
df = pd.read_excel('Global Superstore.xls')    #Reads the data to the variable “df”
df.sample(3)   #shows 3 records from the data
```

```python
# Bar Plot

SalesByPriority = df.groupby(df['Order Priority'])['Sales'].aggregate(np.sum).reset_index()

barGraph = plt.figure()
plt.bar(SalesByPriority['Order Priority'], SalesByPriority['Sales'])
barGraph.axes[0].get_yaxis().set_major_formatter(
tck.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.title("Sales by Order Priority")
plt.xlabel('Order Priority')
plt.ylabel('Sales')
```

```python
# Histogram

bins = 10    # define and alocate a value to a variable, 10 bins here
Hist = plt.figure()
plt.hist((df['Quantity']),bins,facecolor = 'orange')
Hist.axes[0].get_yaxis().set_major_formatter(
tck.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.title("Distribution Of Quantity")
plt.ylabel('Count')
plt.xlabel('Quantity')
```

```python
# Scatter Plot

scatter = plt.figure()
plt.scatter(df['Sales'],df['Profit'], color = 'green')
scatter.axes[0].get_xaxis().set_major_formatter(
tck.FuncFormatter(lambda x, p: format(int(x), ',')))
scatter.axes[0].get_yaxis().set_major_formatter(
tck.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales Vs profit')
```

```python
# Line Graph

SalesByShipMode = df.groupby(df['Ship Mode'])['Sales'].aggregate(np.sum).reset_index()
line = plt.figure()
plt.plot(SalesByShipMode['Ship Mode'],SalesByShipMode['Sales'])
line.axes[0].get_yaxis().set_major_formatter(
tck.FuncFormatter(lambda x, p: format(int(x), ',')))
line.axes[0].set_title('Sales By Ship Mode')
line.axes[0].set_xlabel('Ship Mode')
line.axes[0].set_ylabel('Sales')
```

```python
# Stacked Area Chart

dummy = np.arange(50)
random = np.random.randint(0,10,size=(3,dummy.size))
yrs = 1950+dummy
fig,ax = plt.subplots(figsize=(5,3))
ax.stackplot(yrs,dummy+random,labels = ['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
```
